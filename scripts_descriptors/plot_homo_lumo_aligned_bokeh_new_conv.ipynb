```python
import pandas as pd
import glob
import os
import ntpath

from bokeh.plotting import figure
from bokeh.io import output_file, show, export_png
from bokeh.models import HoverTool, ColumnDataSource, Span
```

```python
babd = pd.read_csv(os.path.join("../output_data/new/all_oldnew_conv.csv"))
```

```python
babd.loc[babd.label=='14080N2_ddec.cif']
```

```python
HER = -4.4
OER = -5.629
VIS_E = 3.2
```

```python
desc = babd['label']
homo = ColumnDataSource(data=dict(x=babd['bandgap_corr'], y=babd['homo_align_corr'], desc=desc))
lumo = ColumnDataSource(data=dict(x=babd['bandgap_corr'], y=babd['lumo_align_corr'], desc=desc))
hover = HoverTool(tooltips=[
    ("index", "$index"),
    ("(x,y)", "(@x,@y)"),
    ('desc', '@desc'),
])
p = figure(plot_width=800, plot_height=400, tools=[hover], x_axis_label='Corrected Band Gap (eV)', y_axis_label='Corrected IP/EA energies (eV)', title="Energy alignment of HOMO/LUMO for CURATED COFs")
p.circle('x', 'y', size=10, source=homo, color='red')
p.circle('x', 'y', size=10, source=lumo, color='blue')
herline = Span(location=HER,dimension='width',line_color='orange',line_dash='dashed',line_width=2,name="HER")
oerline = Span(location=OER,dimension='width',line_color='purple',line_dash='dashed',line_width=2,name="OER")
visline = Span(location=VIS_E,dimension='height',line_color='black',line_dash='dashed',line_width=2,name="VIS")
p.add_layout(herline)
p.add_layout(oerline)
p.add_layout(visline)
output_file("../output_data/new/energy_align_all.html")
show(p)
```

```python
ba_filtered_basic = babd[(babd['bandgap_corr']<VIS_E)&(babd['homo_align_corr']<OER)\
                         &(babd['lumo_align_corr']>HER)]
desc = ba_filtered_basic['label']
homo = ColumnDataSource(data=dict(x=ba_filtered_basic['bandgap_corr'], y=ba_filtered_basic['homo_align_corr'], desc=desc))
lumo = ColumnDataSource(data=dict(x=ba_filtered_basic['bandgap_corr'], y=ba_filtered_basic['lumo_align_corr'], desc=desc))
hover = HoverTool(tooltips=[
    ("index", "$index"),
    ("(x,y)", "(@x,@y)"),
    ('desc', '@desc'),
])
p = figure(plot_width=800, plot_height=400, tools=[hover], x_axis_label='Corrected Band Gap (eV)', y_axis_label='Corrected IP/EA energies (eV)', title="Filtered CURATED COFs for photocatalytic overall water splitting")
p.circle('x', 'y', size=10, source=homo, color='red')
p.circle('x', 'y', size=10, source=lumo, color='blue')
herline = Span(location=HER,dimension='width',line_color='orange',line_dash='dashed',line_width=2,name="HER")
oerline = Span(location=OER,dimension='width',line_color='purple',line_dash='dashed',line_width=2,name="OER")
visline = Span(location=VIS_E,dimension='height',line_color='black',line_dash='dashed',line_width=2,name="VIS")
p.add_layout(herline)
p.add_layout(oerline)
p.add_layout(visline)
output_file("../output_data/new/filtered_energy_align_1_all.html")
# export_png(p, filename = "filtered_energy_align_1_all.png")
show(p)
```

```python
ba_filtered_vis = babd[(babd['bandgap_corr']<VIS_E)&(babd['bandgap_corr']>1.6)]
desc = ba_filtered_vis['label']
homo = ColumnDataSource(data=dict(x=ba_filtered_vis['bandgap_corr'], y=ba_filtered_vis['homo_align_corr'], desc=desc))
lumo = ColumnDataSource(data=dict(x=ba_filtered_vis['bandgap_corr'], y=ba_filtered_vis['lumo_align_corr'], desc=desc))
hover = HoverTool(tooltips=[
    ("index", "$index"),
    ("(x,y)", "(@x,@y)"),
    ('desc', '@desc'),
])
p = figure(plot_width=800, plot_height=400, tools=[hover], x_axis_label='Corrected Band Gap (eV)', y_axis_label='Corrected IP/EA energies (eV)', title="Filtered CURATED COFs for photocatalytic overall water splitting")
p.circle('x', 'y', size=10, source=homo, color='red')
p.circle('x', 'y', size=10, source=lumo, color='blue')
herline = Span(location=HER,dimension='width',line_color='orange',line_dash='dashed',line_width=2,name="HER")
oerline = Span(location=OER,dimension='width',line_color='purple',line_dash='dashed',line_width=2,name="OER")
visline = Span(location=VIS_E,dimension='height',line_color='black',line_dash='dashed',line_width=2,name="VIS")
p.add_layout(herline)
p.add_layout(oerline)
p.add_layout(visline)
output_file("../output_data/new/filtered_energy_vis_all.html")
# export_png(p, filename = "filtered_energy_align_1_all.png")
show(p)
```

```python
ba_filtered_bg = babd[babd['bandgap_corr']<3.2]
ba_filtered_gap_2ev = ba_filtered_bg[(((ba_filtered_bg['homo_align_corr'])>(OER-2.2))\
                                     &((ba_filtered_bg['homo_align_corr'])<(OER)))\
                                     |(((ba_filtered_bg['lumo_align_corr'])<(HER+2.2))\
                                     &((ba_filtered_bg['lumo_align_corr'])>(HER)))]
desc = ba_filtered_gap_2ev['label']
homo = ColumnDataSource(data=dict(x=ba_filtered_gap_2ev['bandgap_corr'], y=ba_filtered_gap_2ev['homo_align_corr'], desc=desc))
lumo = ColumnDataSource(data=dict(x=ba_filtered_gap_2ev['bandgap_corr'], y=ba_filtered_gap_2ev['lumo_align_corr'], desc=desc))
hover = HoverTool(tooltips=[
    ("index", "$index"),
    ("(x,y)", "(@x,@y)"),
    ('desc', '@desc'),
])
p = figure(plot_width=800, plot_height=400, tools=[hover], x_axis_label='Corrected Band Gap (eV)', y_axis_label='Corrected IP/EA energies (eV)', title="Filtered CURATED COFs for photocatalytic HER/OER")
p.circle('x', 'y', size=10, source=homo, color='red')
p.circle('x', 'y', size=10, source=lumo, color='blue')
herline = Span(location=HER,dimension='width',line_color='orange',line_dash='dashed',line_width=2,name="HER")
oerline = Span(location=OER,dimension='width',line_color='purple',line_dash='dashed',line_width=2,name="OER")
visline = Span(location=VIS_E,dimension='height',line_color='black',line_dash='dashed',line_width=2,name="VIS")
p.add_layout(herline)
p.add_layout(oerline)
p.add_layout(visline)
output_file("../output_data/new/filtered_energy_align_2_all.html")
show(p)
```
